```python
# Load the final processed dataset
import pandas as pd
FIGURE_PATH = "../Figures/"

DATA_PATH = "../Dataset/neo_processed.csv"
df_processed = pd.read_csv(DATA_PATH)
print("Loaded shape:", df_processed.shape)
df_processed.head()
```

```python
# Check target column exists and print alternatives if not
possible_targets = ['hazardous_label', 'hazardous', 'is_hazardous', 'pha']
for t in possible_targets:
    if t in df_processed.columns:
        print(f"Found candidate target column: {t}")

# Choose hazardous_label if present, otherwise pick the first candidate found
if 'hazardous_label' in df_processed.columns:
    target_col = 'hazardous_label'
else:
    found = [c for c in possible_targets if c in df_processed.columns]
    if found:
        target_col = found[0]
        print(f"Using alternative target column: {target_col}")
    else:
        print("Available columns:\n", df_processed.columns.tolist())
        raise SystemExit("No target column found. Please ensure the processed dataset contains the target column.")

print(f"Target column selected: {target_col}")
```

```python
# Prepare features and target
X = df_processed.drop(columns=[target_col])
y = df_processed[target_col]

# Drop any original/duplicate label columns if present
for c in ['pha', 'hazardous', 'is_hazardous']:
    if c in X.columns:
        X = X.drop(columns=[c])

# Fill numeric NaNs with median (safety)
num_cols = X.select_dtypes(include=['number']).columns
X[num_cols] = X[num_cols].fillna(X[num_cols].median())

# One-hot encode categorical features (if any)
X = pd.get_dummies(X, drop_first=True)

print("Feature matrix shape after preprocessing:", X.shape)
print("Sample columns:", X.columns[:20].tolist())
```

```python
leaky = []
for col in X.columns:
    matches = (X[col] == y).sum()
    if matches > 0.5 * len(X):   # column matches target more than half the dataset
        leaky.append(col)

print("Leaky columns:", leaky)
```

```python
X_clean = X.drop(columns=leaky)
leaky = []
for col in X_clean.columns:
    matches = (X_clean[col] == y).sum()
    if matches > 0.5 * len(X_clean):   # column matches target more than half the dataset
        leaky.append(col)

print("Leaky columns:", leaky)
```

```python
X_sup =  X_clean.drop(columns=["epoch_cal_2025-11-21.0000000","class_APO","risk_score"])
print(X_sup.columns)
```

```python
# Try DBSCAN for anomaly detection and compare with NASA PHA
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

# Data already scaled as X_unsup, but let's ensure it's standardized
X_dbscan = X_unsup

dbscan = DBSCAN(eps=1.5, min_samples=10)
dbscan_labels = dbscan.fit_predict(X_dbscan)

# DBSCAN labels -1 as outliers/anomalies
df_processed["dbscan_anomaly"] = (dbscan_labels == -1).astype(int)

def nasa_pha(row):
    cond1 = row["moid"] < 0.05
    cond2 = (row["diameter"] > 0.14) or (row["H"] < 22)
    return int(cond1 and cond2)

df_processed["nasa_pha"] = df_processed.apply(nasa_pha, axis=1)

print("\n--- DBSCAN Anomaly vs NASA PHA correlation ---")
print(pd.crosstab(df_processed["dbscan_anomaly"], df_processed["nasa_pha"]))

print("\n% of DBSCAN anomalies that are NASA-PHAs:")
print(df_processed[df_processed["dbscan_anomaly"]==1]["nasa_pha"].mean())
```

```python
# Gaussian Mixture Model (GMM) clustering and anomaly scoring
from sklearn.mixture import GaussianMixture
import numpy as np

# Fit GMM with a reasonable number of components (tunable)
gmm = GaussianMixture(n_components=5, covariance_type='full', random_state=42)
gmm.fit(X_unsup)

# Responsibilities (posterior probs) per component
resp = gmm.predict_proba(X_unsup)
max_resp = resp.max(axis=1)

# Define GMM anomalies as points with low max responsibility (low cluster affinity)
gmm_threshold = np.quantile(max_resp, 0.05)  # bottom 5% affinity as anomalies
gmm_anomaly = (max_resp <= gmm_threshold).astype(int)
df_processed["gmm_anomaly"] = gmm_anomaly

# Compute NASA PHA (if not already present)
def nasa_pha(row):
    cond1 = row["moid"] < 0.05
    cond2 = (row["diameter"] > 0.14) or (row["H"] < 22)
    return int(cond1 and cond2)

if "nasa_pha" not in df_processed.columns:
    df_processed["nasa_pha"] = df_processed.apply(nasa_pha, axis=1)

print("\n--- GMM Anomaly vs NASA PHA correlation ---")
print(pd.crosstab(df_processed["gmm_anomaly"], df_processed["nasa_pha"]))

print("\n% of GMM anomalies that are NASA-PHAs:")
print(df_processed[df_processed["gmm_anomaly"]==1]["nasa_pha"].mean())
```

```python
# Local Outlier Factor (LOF) anomaly detection
from sklearn.neighbors import LocalOutlierFactor

lof = LocalOutlierFactor(n_neighbors=35, contamination=0.03)
# fit_predict returns -1 for anomalies
lof_labels = lof.fit_predict(X_unsup)
df_processed["lof_anomaly"] = (lof_labels == -1).astype(int)

print("\n--- LOF Anomaly vs NASA PHA correlation ---")
print(pd.crosstab(df_processed["lof_anomaly"], df_processed["nasa_pha"]))

print("\n% of LOF anomalies that are NASA-PHAs:")
print(df_processed[df_processed["lof_anomaly"]==1]["nasa_pha"].mean())
```

```python
# IsolationForest tuning: try multiple parameters and pick best by NASA PHA overlap
from sklearn.ensemble import IsolationForest

best_overlap = -1.0
best_params = None
best_labels = None

for contamination in [0.01, 0.02, 0.03, 0.05]:
    for n_estimators in [100, 300]:
        iso = IsolationForest(contamination=contamination, n_estimators=n_estimators, random_state=42)
        iso_labels = iso.fit_predict(X_unsup)
        anomalies = (iso_labels == -1).astype(int)
        overlap = df_processed.assign(tmp=anomalies)[lambda d: d["tmp"]==1]["nasa_pha"].mean()
        if overlap > best_overlap:
            best_overlap = overlap
            best_params = {"contamination": contamination, "n_estimators": n_estimators}
            best_labels = anomalies

df_processed["iso_tuned_anomaly"] = best_labels
print("\nBest IsolationForest params:", best_params)
print("Best NASA PHA overlap:", best_overlap)
print(pd.crosstab(df_processed["iso_tuned_anomaly"], df_processed["nasa_pha"]))
```

```python
# Parameter sweeps for DBSCAN and GMM with summary comparison
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn.mixture import GaussianMixture

results = []

# Ensure NASA PHA exists
def nasa_pha(row):
    cond1 = row["moid"] < 0.05
    cond2 = (row["diameter"] > 0.14) or (row["H"] < 22)
    return int(cond1 and cond2)

if "nasa_pha" not in df_processed.columns:
    df_processed["nasa_pha"] = df_processed.apply(nasa_pha, axis=1)

# Sweep DBSCAN
for eps in [0.8, 1.0, 1.2, 1.5, 2.0]:
    for min_samples in [5, 10, 20, 30]:
        db = DBSCAN(eps=eps, min_samples=min_samples)
        labels = db.fit_predict(X_unsup)
        db_anom = (labels == -1).astype(int)
        overlap = df_processed.assign(tmp=db_anom)[lambda d: d["tmp"]==1]["nasa_pha"].mean()
        results.append({"method":"DBSCAN", "params": {"eps": eps, "min_samples": min_samples}, "overlap": float(overlap)})

# Sweep GMM components and covariance
for n_comp in [3, 5, 8, 12]:
    for cov in ["full", "tied", "diag"]:
        gmm = GaussianMixture(n_components=n_comp, covariance_type=cov, random_state=42)
        gmm.fit(X_unsup)
        resp = gmm.predict_proba(X_unsup)
        max_resp = resp.max(axis=1)
        # mark lowest 3%, 5%, 8% as anomalies and pick best
        for q in [0.03, 0.05, 0.08]:
            thr = np.quantile(max_resp, q)
            anom = (max_resp <= thr).astype(int)
            overlap = df_processed.assign(tmp=anom)[lambda d: d["tmp"]==1]["nasa_pha"].mean()
            results.append({"method":"GMM", "params": {"n_components": n_comp, "covariance_type": cov, "quantile": q}, "overlap": float(overlap)})

# Include previously tuned IsolationForest result if present
if "iso_tuned_anomaly" in df_processed.columns:
    iso_overlap = df_processed[df_processed["iso_tuned_anomaly"]==1]["nasa_pha"].mean()
    results.append({"method":"IsolationForest", "params":"tuned", "overlap": float(iso_overlap)})

summary = pd.DataFrame(results).sort_values("overlap", ascending=False)
print("\n=== Method NASA PHA Overlap Summary (higher is better) ===")
print(summary.head(20))

# Keep the top performing method's anomaly labels in df_processed
best = summary.iloc[0]
print("\nBest method:", best["method"], "params:", best["params"], "overlap:", best["overlap"])
best_method = best["method"]
best_params = best["params"]
best_labels = None

if best_method == "DBSCAN":
    db = DBSCAN(**best_params)
    best_labels = (db.fit_predict(X_unsup) == -1).astype(int)
elif best_method == "GMM":
    gmm = GaussianMixture(n_components=best_params["n_components"], covariance_type=best_params["covariance_type"], random_state=42)
    gmm.fit(X_unsup)
    resp = gmm.predict_proba(X_unsup)
    max_resp = resp.max(axis=1)
    thr = np.quantile(max_resp, best_params["quantile"])
    best_labels = (max_resp <= thr).astype(int)
elif best_method == "IsolationForest":
    best_labels = df_processed["iso_tuned_anomaly"].values.astype(int)

df_processed["unsup_best_anomaly"] = best_labels
print(pd.crosstab(df_processed["unsup_best_anomaly"], df_processed["nasa_pha"]))
print("% of best-method anomalies that are NASA-PHAs:", df_processed[df_processed["unsup_best_anomaly"]==1]["nasa_pha"].mean())
```
